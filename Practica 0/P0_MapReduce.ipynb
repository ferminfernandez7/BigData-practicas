Debes completar los recuadros de Markdown que indican **"TU RESPUESTA AQUÍ"** para producir la salida esperada. Además, debes incluir una **breve explicación** y un **dibujo** de tu solución MapReduce.

Para los dibujos, utiliza un estilo similar al que empleamos en el **Tema 2** con el ejemplo de *Word Count*:

Es importante que describas bien cómo funcionarán `map` y `reduce`. En el ejemplo de *Word Count*: (1) `map` emite cada palabra como **clave** de salida y asigna un `1` como **valor** de salida para contar esa palabra; (2) `reduce` agrega la lista de unos sumando sus elementos: `lambda x, y: x + y`, realizando la reducción por pares (de dos elementos cada vez).

- **Comprensión**: las explicaciones deben demostrar que entiendes bien el concepto de MapReduce.

- **Eficiencia**: por ejemplo, ¿tu solución **escalaría** bien a ficheros de entrada muy grandes?

## Ejercicio 1. Histograma de repeticiones de palabras

Debes diseñar las funciones `map` y `reduce` para obtener un **histograma de repeticiones de palabras** en uno o varios documentos de texto. Es decir: cuántas palabras se repiten **una** vez, cuántas se repiten **dos** veces, cuántas se repiten **tres** veces, etc.

**Entrada**: uno o varios documentos con texto plano. Puedes suponer que cada línea (o documento) tiene un identificador aleatorio que se usará como clave de entrada en la fase `map`, y que los valores de entrada son las líneas de texto o documentos completos. Por ejemplo:

    (1, "Welcome to the Big")
    (2, "World of Big Big")
    (3, "Data Welcome World bye")

**Salida**: tu salida debe indicar cuántas palabras se repiten 1, 2, etc. veces, donde el número de repeticiones es la clave de salida y el número total de palabras con ese número de repeticiones es el valor (es decir, `(número de repeticiones, número de palabras)`). Para la entrada anterior, la salida sería:

    (1 vez - 5 palabras)
    (2 veces - 2 palabras)
    (3 veces - 1 palabra)

**Pista**: puedes encadenar tantos procesos MapReduce como quieras; por ejemplo, podrías aplicar primero la solución de *Word Count* y usar su salida como entrada de una segunda etapa.


TU RESPUESTA AQUI

ESQUEMA:
A) Solución de Word Count
- Map: (id, texto) > (palabra, 1)
- Shuffle: (palabra, 1) > (palabra, [1,1...])
- Reduce: (palabra, [1,1...]) > (palabra, apariciones)

B) Histograma
- Map: (palabra, apariciones) > (apariciones, 1)
- Shuffle: (apariciones, 1) > (apariciones, [1,1...])
- Reduce: (apariciones, [1,1...]) > (apariciones, num_palabras)

EXPLICACIÓN:
- Map: Emite el número de apariciones como nueva clave y un 1 como valor
- Shuffle: Agrupa los valores por la nueva clave (apariciones)
- Reduce: Suma la lista de unos, para obtener num_palabras como valor

## Ejercicio 6. URLs enlazadas mutuamente

Obtén una lista de **URLs enlazadas mutuamente**. Es decir, debes encontrar qué URLs se enlazan entre sí (de forma bidireccional) en el contenido de sus páginas.

Debes diseñar una solución MapReduce suponiendo que la entrada es un fichero grande que contiene, para cada URL, la lista de URLs que aparecen enlazadas en su contenido.

**Entrada**: un fichero con el siguiente formato: URL -> lista de URLs

    URL_1 -> URL_2 URL_4
    URL_2 -> URL_1 URL_3 URL_4 URL_5 URL_6
    URL_3 -> URL_2 URL_5 URL_6
    URL_4 -> URL_2 URL_3 URL_5
    URL_5 -> URL_2 URL_3 URL_4
    URL_6 -> URL_2 URL_4

**Salida**: (par de URLs enlazadas mutuamente)

    (URL_1, URL_2)
    (URL_2, URL_3)
    (URL_2, URL_4)
    (URL_2, URL_5)
    (URL_2, URL_6)
    (URL_3, URL_5)
    (URL_4, URL_5)


TU RESPUESTA AQUÍ

ESQUEMA:
- Map: (url_origen, lista_destinos) > ((url_menor, url_mayor), url_origen)
- Shuffle: ((url_menor, url_mayor), url_origen) > ((url_menor, url_mayor), [lista_origenes...])
- Reduce: ((url_menor, url_mayor), [lista_origenes...]) > (url_A, url_B)

EXPLICACIÓN:
- Map: Para cada enlace saliente, genera una clave con el par de URLs ordenado alfabéticamente y emite el origen como valor.
- Shuffle: Agrupa los orígenes de los enlaces bajo la misma clave de par de URLs.
- Reduce: Comprueba si la lista de valores contiene ambas URLs (o tiene tamaño 2), lo que indica que el enlace es bidireccional, y emite el par.

## Ejercicio 7. Encontrar amigos en común entre pares de amigos

Debes diseñar las funciones `map` y `reduce` para obtener, para cada **par de amigos**, la lista de **amigos en común**. Es decir, para cada dos personas que son amigas (suponiendo que la amistad es bidireccional), debes encontrar qué amistades comparten.

Como entrada recibes un fichero que contiene la lista de amigos de cada persona. Usando `map` y `reduce`, debes obtener para cada par de amigos la lista de amigos comunes.

**Entrada**: un fichero con el siguiente formato: Persona -> lista de amigos

EXPLICACIÓN:
- Map: Genera una clave con el par de amigos (ordenado alfabéticamente) y envía la lista completa de amigos como valor.
- Shuffle: Agrupa las listas de amigos provenientes de ambas personas bajo la misma clave.

## Ejercicio 9. Búsqueda del vecino más cercano

Debes diseñar las funciones `map` y `reduce` para encontrar el índice del **vecino más cercano** (nearest neighbor) de un `vector` de entrada dentro de un conjunto de vectores. Puedes suponer que el `vector` de entrada está disponible en todos los nodos de trabajo.

Considera la **distancia euclídea** para calcular la similitud. La distancia euclídea entre dos vectores `a`, `b` de longitud `N` se calcula como:

$$ dist(a, b) = \sqrt{ \sum^{N}_{i=1}{ (a[i] - b[i])^2 } } $$

**Entrada**: un conjunto de vectores como pares clave-valor (la clave es el índice y el valor es el vector) y un `vector` para el que debes encontrar su vecino más cercano.

    Conjunto de vectores:
        (0, (1, 2))
        (1, (1, 1))
        (2, (2, 1))
        (3, (0, 1))

    Vector de entrada: (0, 0.5)

**Salida**: índice del vector más similar al de entrada (el que tenga la menor distancia).

    Índice: 3 (corresponde al vector (0, 1))


TU RESPUESTA AQUÍ

ESQUEMA:
- Map: (id_vector, vector) > (1, (id_vector, distancia))
- Shuffle: (1, (id_vector, distancia)) > (1, [(id_vector, distancia)...])
- Reduce: (1, [(id_vector, distancia)...]) > (id_vecino_mas_cercano, distancia_minima)

EXPLICACIÓN:
- Map: Calcula la distancia entre el vector actual y el vector de consulta (Q). Emite una clave constante (1) para enviar todos los resultados al mismo sitio.
- Shuffle: Agrupa todas las tuplas de (id, distancia) bajo la misma clave.
- Reduce: Recorre la lista de valores para encontrar la distancia mínima y devuelve el ID del vector asociado.

## Ejercicio 10. Búsqueda del vecino más cercano para un conjunto de vectores

Debes diseñar las funciones `map` y `reduce` para encontrar, para cada vector en una lista de `vectors` de entrada, el índice del **vecino más cercano** dentro de un conjunto de vectores. Puedes suponer que la lista de vectores de entrada está disponible para todos los nodos de trabajo.

**Entrada**: un conjunto de vectores en pares clave-valor (la clave es el índice y el valor es el vector) y una lista de vectores para los que debes encontrar su vecino más cercano (con el mismo formato).

    Conjunto de vectores:
        (0, (1, 2))
        (1, (1, 1))
        (2, (2, 1))
        (3, (0, 1))

    Vectores de entrada:
        (0, (0, 0.5))
        (1, (1, 1))

**Salida**: índices de los vectores más similares a los de entrada (los que tengan la menor distancia). El resultado se presentará como una tupla con el índice del vector de entrada y el índice de su vecino más cercano correspondiente.

    Índices:
        (0, 3) (el vector de entrada 0 tiene como vecino más cercano el vector (0, 1))
        (1, 1) (el vector de entrada 1 tiene como vecino más cercano el vector (1, 1))

**Consejo**: puedes resolver el problema con un único MapReduce.


TU RESPUESTA AQUÍ

ESQUEMA:
- Configuración: Cargar lista de Queries [Q1, Q2...] en memoria de todos los mappers.
- Map: (id_v, vector_v) > iterar Queries > (id_query, (id_v, distancia))
- Shuffle: (id_query, (id_v, distancia)) > (id_query, [(id_v, distancia)...])
- Reduce: (id_query, [(id_v, distancia)...]) > (id_query, id_v_mas_cercano)

EXPLICACIÓN:
- Map: Cada mapper tiene la lista completa de Queries en memoria. Para cada vector de la base de datos (V), calcula la distancia contra TODAS las queries y emite un resultado para cada una. La clave es el ID de la Query.
- Shuffle: Agrupa todos los candidatos encontrados en la base de datos para una Query específica.
- Reduce: Recibe todas las distancias calculadas para una Query concreta, busca el mínimo y devuelve qué vector de la base de datos es el más cercano.
